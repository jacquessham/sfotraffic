```python
import pandas as pd
```

```python
df = pd.read_csv('Air_Traffic_Passenger_Statistics.csv')
```

```python
df
```

```python
df = df.drop(['Operating Airline','Operating Airline IATA Code'],axis=1)
df = df.rename(columns={'Activity Period':'date',
                   'Published Airline':'pub_airlines','Published Airline IATA Code':'pub_code',
                   'GEO Summary':'geo_summ', 'GEO Region': 'geo_region', 'Activity Type Code':'type',
                   'Price Category Code':'price', 'Terminal': 'terminal', 'Boarding Area': 'boarding_area',
                   'Passenger Count': 'pax_count'
                  })
```

```python
df.head()
```

```python
df['pub_airlines'].unique()
```

```python
temp = df['pub_airlines'].replace('United Airlines - Pre 07/01/2013','United Airlines')
temp.unique()
```

```python
df['pub_airlines'] = df['pub_airlines'].replace('United Airlines - Pre 07/01/2013','United Airlines')
df['pub_airlines'] = df['pub_airlines'].replace('Emirates ','Emirates')
df['price'] = df['price'].replace('Other', 'Full Service')
```

```python
df['pub_airlines'].unique()
```

```python
df.loc[(df.pub_airlines=='Air China')& (df.price!='Full Service')]
```

```python
df.loc[df.pub_airlines=='Air China', 'price'] = 'Full Service'
```

```python
df.loc[(df.pub_airlines=='Air China')& (df.price!='Full Service')]
```

```python
full_airlines = ['Air China', 'Air India Limited', 'Air New Zealand', 'Air Pacific Limited dba Fiji Airways',
                      'Emirates', 'United Airlines', 'Virgin America', 'Volaris Airlines', 'Delta Air Lines',
                      'US Airways']
for airline in full_airlines:
    df.loc[df.pub_airlines==airline, 'price'] = 'Full Service'
```

```python
lcc_airlines = ['XL Airways France', 'WOW Air', 'WestJet Airlines']
for airline in lcc_airlines:
    df.loc[df.pub_airlines==airline, 'price'] = 'Low Fare'
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m')
```
